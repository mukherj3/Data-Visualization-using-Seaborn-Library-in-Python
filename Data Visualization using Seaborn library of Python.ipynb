# Distribution Plots

```python
penguins_data = penguins_data.dropna()
```

# Univariate histograms

```python
sns.displot(penguins_data, x="flipper_length_mm")
```

```python
sns.displot(penguins_data, x="flipper_length_mm", binwidth=5)
```

```python
sns.displot(penguins_data, x="flipper_length_mm", bins=100)
```

# Distribution Plots, conditioning on other variables

# Conditioning histograms on other variable: Step

```python
sns.displot(penguins_data, x="flipper_length_mm", hue="species", element="step")   ##conditioning on other variables: step
```

# Conditioning histograms on other variable: Stack

```python
sns.displot(penguins_data, x="flipper_length_mm", hue="species", multiple="stack")  ####conditioning on other variable: stack
```

```python
sns.displot(penguins_data, x="flipper_length_mm", hue="sex", multiple="dodge")  ##using separate bars for categorical variable
```

```python
sns.displot(penguins_data.dropna(), x="flipper_length_mm", col="sex", multiple="dodge")
```

```python
penguins_data.isnull().values.any()
```

# Histograms with multiple facets

```python
sns.displot(penguins_data, x="flipper_length_mm", col="sex")
```

# Kernel Density Estimation


```python
sns.displot(penguins_data, x="flipper_length_mm", kind="kde")
```

# Changing the bandwidth of KDE

```python
sns.displot(penguins_data, x="flipper_length_mm", kind="kde", bw_adjust=0.75)
```

```python
sns.displot(penguins_data, x="flipper_length_mm", hue="species", kind="kde")
```

# KDE conditioned on another dimension

```python
sns.displot(penguins_data, x="flipper_length_mm", hue="species", kind="kde", fill=True)
```

# KDE vs Histograms

```python
sns.displot(penguins_data, x="flipper_length_mm", kde=True)
```
